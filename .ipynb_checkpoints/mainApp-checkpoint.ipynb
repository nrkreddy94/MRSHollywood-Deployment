```python
import pickle
import numpy as np
import pandas as pd
import math
```

```python
movie_features_index_df=pd.read_csv("movie_user_rating_index.csv")
movie_features_index_df
```

```python
movie_user_rating_pivottable=pd.read_csv("movie_user_rating_pivottable.csv",encoding ='utf-8',compression='gzip')
movie_user_rating_pivottable
```

```python
model_knn = pickle.load( open( "collaborativeFiltering_model.pkl", "rb" ) )
```

```python
model_classifier = pickle.load( open( "CollaborativeFiltering_NLP_model.pkl", "rb" ) )
```

```python
vectorizer = pickle.load(open("CollaborativeFiltering_NLP_vectorizer.pkl", 'rb'))  
```

```python

def getMovieIndex(name):
    name=name.strip()
    test_vector= vectorizer.transform([name])
    predicted = model_classifier.predict(test_vector)
    print("{0} : {1}".format(predicted[0], movie_features_index_df.title[predicted[0]]))
    return predicted[0]
```

```python
def getRecomendedMoviesByIndex(queryIndex):
    distances, indices = model_knn.kneighbors(movie_user_rating_pivottable.iloc[queryIndex,:].values.reshape(1, -1), n_neighbors = 21)
    return distances,indices
```

```python

def recommendedMovies(name):
    movieList=list()
    query_index=getMovieIndex(name)
    distances, indices =getRecomendedMoviesByIndex(query_index)
    for i in range(0, len(distances.flatten())):
        movieList.append(movie_features_index_df.title[indices.flatten()[i]])
    return movieList

    
```

```python
name="Iron Man"
movieList=recommendedMovies(name)
movieList
```
